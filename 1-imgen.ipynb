```python
import keras
import keras_cv
```

```python
stable_diffusion = keras_cv.models.StableDiffusion(img_height=256, img_width=256)
stable_diffusion
```

```python
images = stable_diffusion.generate_image(stable_diffusion.encode_text("scorpion constellation"), batch_size=2, num_steps=200)
```

```python
from matplotlib import pyplot as plt
for image in images:
    plt.imshow(image)
    plt.show()
```
